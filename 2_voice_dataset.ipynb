```python
class Hello:
    def __enter__(self):
        # ÏÇ¨Ïö©Ìï† ÏûêÏõêÏùÑ Í∞ÄÏ†∏Ïò§Í±∞ÎÇò ÎßåÎì†Îã§(Ìï∏Îì§Îü¨ Îì±)
        print('enter...')
        return self # Î∞òÌôòÍ∞íÏù¥ ÏûàÏñ¥Ïïº VARIABLEÎ•º Î∏îÎ°ùÎÇ¥ÏóêÏÑú ÏÇ¨Ïö©Ìï† Ïàò ÏûàÎã§

    def sayHello(self, name):
        # ÏûêÏõêÏùÑ ÏÇ¨Ïö©ÌïúÎã§. ex) Ïù∏ÏÇ¨ÌïúÎã§
        print('hello ' + name)

    def __exit__(self, exc_type, exc_val, exc_tb):
        # ÎßàÏßÄÎßâ Ï≤òÎ¶¨Î•º ÌïúÎã§(ÏûêÏõêÎ∞òÎÇ© Îì±)
        print('exit...')

# h = Hello()

with Hello() as h:
   h.sayHello('obama')
```
